```python
df={'Student ID':40, 'Degree of difficulty of assignments':50,'Solves doubts willingly':90, 'Structuring of the course':45}
names=list(df.keys())
values=list(df.values())
y=[100000]
plt.pie(values, labels=names,autopct='%0.2f%%')
plt.pie(y,radius=0.4,colors='b')
plt.show()
```

```python
df={'Student ID':20,'Well versed with the subject':30,'Structuring of the course':15}
names=list(df.keys())
values=list(df.values())
plt.pie(values, labels=names, autopct='%0.2f%%',explode=[0,0,0.2])
plt.show()
```

```python
df={'Student ID':12,'Solves doubts willingly':35,'Structuring of the course':54}
names=list(df.keys())
values=list(df.values())
fig=plt.figure(figsize=(5,5))
plt.bar(names,values,color='orange')
plt.title('Bar plot Demo')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(['Bar'],loc='upper right')
```
